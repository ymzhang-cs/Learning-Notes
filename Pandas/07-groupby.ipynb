```python
# 分组聚合
```

```python
df = pd.read_csv('datasets/directory.csv')
df
```

```python
grouped = df.groupby(by='Country')
```

```python
grouped.size()
```

```python
grouped.count()
```

```python
grouped['Brand']
```

```python
grouped["Brand"].count()
```

```python
# 多组聚合
df.groupby(by=['Country', 'State/Province']).count()['Brand']
```

```python
df.groupby(by=['Country', 'State/Province']).count()['Brand']['CN']['11']
```

```python
# 对Series进行聚合
df['Brand'].groupby(by=[df['Country'], df['State/Province']]).count()
```

```python
# 注意到，上面的返回值是一个Series，而不是DataFrame
# 如果想要返回DataFrame，可以在索引时使用列表
df.groupby(by=[df['Country'], df['State/Province']])[['Brand']].count()
```
